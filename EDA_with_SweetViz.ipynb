```
# use penguins dataset
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
X = df.drop('species',axis=1)
Y = df['species']
X
```

```
import sweetviz as sv
analysis_report = sv.analyze(df)# perform EDA analysis
analysis_report.show_html('penguins_dataset_EDA.html',open_browser=True)
```

```
# use Ipython to display the HTML
import IPython
IPython.display.HTML('penguins_dataset_EDA.html')
```

From the above visualization, we can be able to tell valuable features such as distinct values and their distribution. From this, we can be able to get a better feel of our data. For example, in the sex column, we can deduce that the males are very slightly more than the females. Almost equal. From the species and island columns, we can be able to deduce the category with the largest share. For columns with lots of distinct variables, they are binned and displayed as a histogram. From the histogram, we can then infer skewness or even distribution.
For example, in the bill length column, very few have longer bill length.

```
# data splitting for comparison purposes
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3)
```

```
# perform comparison using sweetviz
comparison_report = sv.compare([X_train,'Train'],[X_test,'Test'])# perform EDA analysis
comparison_report.show_html('penguins_dataset_comparison_report.html',open_browser=False)
IPython.display.HTML('penguins_dataset_comparison_report.html')
```

The above comparison report is almost the same as the analysis report but it compares the training set(blue) to the test set(orange). This is vital in judging if the train test split is evenly distributed. This can inform you if the train test split is even and if the test set is a good representation of the train set
