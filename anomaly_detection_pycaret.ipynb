Setup the anomaly detection dataset

```
from pycaret.datasets import get_data
dataset = get_data('anomaly')
from pycaret.anomaly import *
setup = setup(dataset)
```

# Create Anomaly detection model

```
detection_model = create_model('iforest') # more detection algorithms at pycaret.org/anomaly-detection/ eg knn, abod, histogram, lof
detection_model
plot_model(detection_model)
```

# Create model and perform mock prediction

```
predictions = predict_model(detection_model, data=dataset) # at this point, make predictions on test set
predictions
```
