```
import klib
import pandas as pd

df = pd.read_csv('https://github.com/akanz1/klib/raw/main/examples/NFL_DATASET.csv')
# provide high level overview of missing data
klib.missingval_plot(df)
```

```
clean_df = klib.data_cleaning(df)# drop duplicate rows and single valued columns
```

```
#correlation plot
klib.corr_plot(clean_df,annot=False)
```

```
# split correlation- Display only positive(pos) or negative(neg)
klib.corr_plot(clean_df,annot=False,split='pos')
```

```
# correlation for a particular column
klib.corr_plot(clean_df,annot=False, target='Touchdown')
```

From the above plot, it can be seen that the sp feature is the one with the highest positive correlation to Touchdown and the yrdln is the feature with the highest negative correlation. This can inform you better on the nature of your data.

```
# correlation matrix
klib.corr_mat(clean_df)
```

```
# distribution plot
klib.dist_plot(clean_df['Win_Prob'])
```

From the plot above, the distribution is almost even with the highest density at around 0.5 Win_Prob

```
# categorical plot
klib.cat_plot(df,figsize=(60,30))
```

```
# save plot tp pdf. Comes in handy when you would wish to share reports
plot = klib.corr_plot(clean_df,annot=False,split='pos')
plot.figure.savefig('pos_corr.pdf')
```
